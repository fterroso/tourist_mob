```python
from pyrosm.data import sources

from pyrosm import OSM
from pyrosm import get_data

import matplotlib.pyplot as plt
%matplotlib inline
```

Spain is one of the available data sources

```python
print(sources.europe.available)
```

```python
fp= get_data('spain')
osm = OSM(fp)
```

```python
osm
```

## Hotels in Spain

```python
hotel_custom_filter_ = {'tourism': ['hotel', 'guest_house', 'hostel', 'motel'], 'building': ['hotel']}
```

```python
hotels_osm = osm.get_data_by_custom_criteria(custom_filter=hotel_custom_filter_,
                                        # Keep data matching the criteria above
                                        filter_type="keep",
                                        # Keep nodes (point data)    
                                        keep_nodes=True, 
                                        keep_ways=False, 
                                        keep_relations=False)
```

```python
hotels_osm
```

```python
hotels_osm.columns
```

```python
hotels_osm['tourism'].unique()
```

```python
hotels_osm.shape
```

```python
hotels_osm.plot('tourism', figsize=(20,10), legend=True);
```

```python
hotels_osm.to_csv('hotels.csv')
```
